> Sea $x \in I = [-1, 1]$, se desea resolver la siguiente EDO que llamaremos $(A)$:
> \begin{cases} & u''(x) + \frac{\pi^{2}}{4}u(x) = 0 \\ & u(1) = u(-1) = 0, u(0) = 1\\ \end{cases}
>
> #### $\cdot$) Verificar que $B(x) = \cos \left( \frac{\pi}{2}x \right) $ resuelve (A).
> Para la demostracion se sustituyo la funcion $b(x)$ en la EDO, siguiendo asi los siguientes calculos el desarrollo:
> \begin{align*} u(x) &:= B(x) \\ \implies  u''(x) + \frac{\pi^2}{4}u(x) = 0 & \iff B''(x) + \frac{\pi^2}{4}B(x) = 0 & /  B(x) = \cos(\frac{\pi}{2}x)\\ \implies B''(x) + \frac{\pi^2}{4}B(x) = 0 & \iff \left( \cos(\frac{\pi}{2}x)\right) '' + \frac{\pi^2}{4}\cos(\frac{\pi}{2}x) = 0 & \text{(Regla de la cadena)}\\ \iff  B''(x) + \frac{\pi^2}{4}B(x) = 0 & \iff \left(-\frac{\pi}{2}\sin(\frac{\pi}{2}x)\right)' + \frac{\pi^2}{4}\cos(\frac{\pi}{2}x) = 0 & \text{(Regla de la cadena)} \\ \iff  B''(x) + \frac{\pi^2}{4}B(x) = 0 & \iff -\frac{\pi^2}{4}\cos(\frac{\pi}{2}x) + \frac{\pi^2}{4}\cos(\frac{\pi}{2}x) = 0 \\ \end{align*}

 > ### $1)\ $ Generar colección de datos $D$
 > Sea $N = 100$. Tome al azar ahora $N$ puntos $x_{j} \in [−1, 1], {j = 1,...,N}$ diferentes. Estos puntos no serán cambiados en lo que sigue y serán denotados como $D = \{x_{j}\}^{N}_{j=1}$.  Grafique estos puntos en la línea $[−1, 1]$, verificando que están bien distribuidos.

> Para el desarrollo del método se consideró una colección con $N = 100$ valores que fueron generados de manera aleatoria, donde todos los valores $x_{j} \in [-1, 1]$, dicho de otra forma, $\{\forall x \in D, |x| \leq 1\}$. Los valores se generaron basándonos en el siguiente código asociado que genera $100$ puntos aleatorios en el intervalo $[-1, 1]$, donde se asociaron una función $N(cantidad)$ para enumerar estos valores en caso de ser necesario y $xj(N, intervalo)$ que genera el conjunto recibiendo la cantidad $N$ de datos y el intervalo de valores.

````python
# ============================= Tarea 3, Parte 2.1 ==============================
########################### Generar coleccion D #################################
n = 100
conjunto = [-1, 1]


def N(cantidad):
    NN = []
    for i in range(cantidad):
        NN += [i+1]
    return NN


def xj(N, intervalo):
    sup = intervalo[1]
    inf = intervalo[0]
    lista_datos_aleatorios = []
    for i in range(N):
        lista_datos_aleatorios += [float(np.random.uniform(inf, sup))]
    return lista_datos_aleatorios


xji = xj(n, conjunto)
Nn = N(n)


# Ver datos generados
# print('La coleccion D con N = '+str(len(xji))+' datos tiene la formada de: \n D = '+str(xji))
########################### Generar coleccion D #################################
# ============================= Tarea 3, Parte 2.1 ==============================
````

````python
# ============================= Tarea 3, Parte 2.1 ==============================
########################### Generar coleccion D #################################
ejen = []
cotaU = []
cotaL = []
for i in range(n):
    ejen += [0]
    cotaU += [1]
    cotaL += [-1]


plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(xji, ejen, label='Valores para $x_{j}$', color='green')
plt.title("Gráfico de distribución para los " + str(n) + " valores aleatorios \n entre $-1$ y $1$")
plt.xlabel("Rango de $N$-ésimo valor para $x_{j} \in [-1, 1]$")
plt.legend(loc='lower right')

plt.subplot(1, 2, 2)
plt.plot(Nn, xji, '*', color='black')
plt.plot(Nn, xji, label="Valores aleatorios", color='green')
plt.plot(Nn, cotaU, '--', color='0.3')
plt.plot(Nn, cotaL, '--', color='0.3')
plt.title("Valor de los $N = " + str(n) + "$ datos aleatorios \n entre $-1$ y $1$")
plt.xlabel("$N$-ésimo valor")
plt.ylabel('Valor de $x_{j}$')
plt.legend(loc='lower right')

plt.tight_layout()
plt.show()
########################### Generar coleccion D #################################
# ============================= Tarea 3, Parte 2.1 ==============================
````

> ### $2)$ Definir la función de activación $\sigma$, la realización de la red neuronal $\mathcal{R}(\phi)(x)$ y la función de costos $C(\phi)$ utilizadas.
> Sea $\sigma(s) = \sin(s)$. Considere una red neuronal $\Phi = (w_1, w_2, b_1, b_2)$ tal que su realización sea una función de clase $\mathcal{C^{2}$ dada por $R(\Phi) : \mathbb{R} \to \mathbb{R}$ como sigue:
> \begin{equation*} R(\Phi)(x) = w_2\sigma(w_1x + b_1) + b_2. \end{equation*}
> Considere además las funciones de costo:
> \begin{align*} C_1(u) &= \frac{1}{N} \sum_{j=1}^{N} \left(u''(x_j) + \frac{\pi^2}{4}u(x_j)^2\right) \\
> C_2(u) &= \frac{1}{3}\left(u^2(-1) + u^2(1) + (u(0)-1)^2\right) \end{align*}
> Las funciones $C_1$ y $C_2$ miden, respectivamente: 1) si $u$ es solución (depende de $D$ acá) y 2) si satisface las condiciones de borde. A
> partir de estas funciones de costo auxiliares, defina la siguiente función de costo principal:
> \begin{equation*}C(w_1, w_2, b_1, b_2) = \frac{1}{2}\left(C_1(u) + C_2(u)\right)(R(\Phi)) \end{equation*}
> Calcule las funciones $\sigma''$ y $\mathcal{R}(\Phi)''$, para luego calcular $\nabla C(w_{1}, w_{2}, b_{1}, b_{2})$

> Lo siguiente fue definir la función de costo principal y funciones asociadas como $\sigma''$ o derivada de orden 2 para la función de activación, $\sigma$, $\mathcal{R}(\Phi)''$ o derivada de orden 2 para la realización $\mathcal{R}(\Phi)$ y, $\nabla C(w_{1}, w_{2}, b_{1}, b_{2})$ o gradiente de la función de costos, además de funciones agregadas de manera externa para la realización. Asi, la definición de estas funciones fue dada por la siguiente organización:
>
> Función de activación o $\sigma(s)$:

````python
# ============================= Tarea 2, Parte 2.b =============================
###################### Funcion de activacion sigma #############################
def sigma(s, orden):
    if orden == 0:
        #print('dsen = sen(s) <=> dsen = sen('+str(s)+') = '+str(np.sin(s)))
        return np.sin(s)
    elif orden == 1:
        #print('dsen = cos(s) <=> dsen = cos('+str(s)+') = '+str(np.cos(s)))
        return np.cos(s)
    elif orden == 2:
        #print('dsen = -sen(s) <=> dsen = -sen('+str(s)+') = '+str(-1*np.sin(s)))
        return -np.sin(s)
    elif orden == 3:
        #print('dsen = -cos(s) <=> dsen = -cos('+str(s)+') = '+str(-1*np.cos(s)))
        return -np.cos(s)
###################### Funcion de activacion sigma #############################
# ============================= Tarea 2, Parte 2.b =============================
````

> Funcion realizacion de la red neuronal o $\mathcal{R}(\Phi)$.

````python
# ============================= Tarea 2, Parte 2.b =============================
######################### Funcion de realizacion ###############################
def R_phi(Phi, x, orden):
    w1, w2, b1, b2 = Phi
    en_sigma = b1+w1*x
    if orden == 0:
        return w2*sigma(en_sigma, 0) + b2
    elif orden == 1:
        return w1*w2*sigma(en_sigma, 1)
    elif orden == 2:
        return w2*sigma(en_sigma, 2)*w1*w1
######################### Funcion de realizacion ###############################
# ============================= Tarea 2, Parte 2.b =============================
````

> Funcion gradiente de la realizacion o $\nabla \mathcal{R}^{\left[orden\right]}(\Phi)$ (extra agregado).

````python
# ============================= Tarea 2, Parte 2.b =============================
################ Funcion del gradiente de la realizacion #######################
def Gradiente_Rphi(Phi, x, nnabla):
    w1, w2, b1, b2 = Phi
    en_sigma = b1 + w1*x
    if nnabla == 1:
        dw1 = w2*x*sigma(en_sigma, 1)
        dw2 = sigma(en_sigma, 0)
        db1 = w2*sigma(en_sigma, 1)
        db2 = 1
        grad1 = np.array([dw1, dw2, db1, db2])
        return grad1
    elif nnabla == 2:
        d2w1 = 2*w1*w2*(sigma(en_sigma,2)) + w2*x*(sigma(en_sigma, 3))*(w1**2)
        d2w2 = sigma(en_sigma, 2)*(w1**2)
        d2b1 = w2*sigma(en_sigma, 3)*(w1**2)
        d2b2 = 0
        grad2 = np.array([d2w1, d2w2, d2b1, d2b2])
        return grad2
################ Funcion del gradiente de la realizacion #######################
# ============================= Tarea 2, Parte 2.b =============================
````

> Funcion de costos o $C(w_{1}, w_{2}, b_{1}, b_{2})$, para esta funcion, se definieron las funciones auxiliares dentro de estas ultimas.

````python
# ============================= Tarea 2, Parte 2.b =============================
############################ Funcion de costos #################################
def C_phi(Phi, D):
    N = len(D)
    def Coste_C1(D):
        N = len(D)
        pi2 = ((np.pi ** 2)/4)
        C1 = 0
        for i in range(N):
            R2 = R_phi(Phi, D[i], 2)
            R = R_phi(Phi, D[i], 0)
            C1 += (R2+pi2*R)**2
            #print(i+1)
        return C1
    def Coste_C2(Phi):
        U_1 = ((R_phi(Phi, -1, 0))**2)
        U1 = ((R_phi(Phi, 1, 0))**2)
        U0 = ((R_phi(Phi, 0, 0)-1)**2)
        C2 = (U_1+U1+U0)
        return C2
    Costo1 = (1/N)*Coste_C1(D)
    Costo2 = (1/3)*Coste_C2(Phi)
    Costo = (1/2)*(Costo1+Costo2)
    return Costo
############################ Funcion de costos #################################
# ============================= Tarea 2, Parte 2.b =============================
````

> Finalmente, funcion gradiente del costo o $\nabla C(w_{1}, w_{2}, b_{1}, b_{2})$, la cual fue seccionada en tres funciones que interactuan mutuamente.

````python
# ============================= Tarea 2, Parte 2.b =============================
######################## Funcion gradiente de costos ###########################
def Parcial(Phi, D, costo):
    if costo == 1:
        R2 = R_phi(Phi, D, 2)
        R = R_phi(Phi, D, 0)
        pi2 = ((np.pi ** 2) / 4)
        CR = 2*(R2+pi2*R)
        Cc2 = Gradiente_Rphi(Phi, D, 2)
        Cc = Gradiente_Rphi(Phi, D, 1)
        parcial = (Cc2+pi2*Cc)
        dC1 = CR*parcial
        grad1 = (dC1[0], dC1[1], dC1[2], dC1[3])
        return grad1
    elif costo == 2:
        CU_1 = 2*(R_phi(Phi, -1, 0))
        CU1 = 2*(R_phi(Phi, 1, 0))
        CU0 = 2*(R_phi(Phi, 0, 0)-1)
        dCU_1 = Gradiente_Rphi(Phi, -1, 1)
        dCU1 = Gradiente_Rphi(Phi, 1, 1)
        dCU0 = Gradiente_Rphi(Phi, 0, 1)
        parcial1 = CU_1*dCU_1
        parcial2 = CU1*dCU1
        parcial3 = CU0*dCU0
        dC2 = parcial1+parcial2+parcial3
        grad2 = ((1/3)*dC2[0], (1/3)*dC2[1], (1/3)*dC2[2], (1/3)*dC2[3])
        return grad2


def derivadaParcial(Phi,D,gradn):
    N = len(D)+1
    if N==1:
        return gradn
    else:
        datos = D[1:N]
        return derivadaParcial(Phi, datos, gradn = gradn + [Parcial(Phi, D[0], 1)])


def Grad_C1(Phi,D):
    n = len(D)
    datos = derivadaParcial(Phi,D,gradn=[])
    dw1 = []
    dw2 = []
    db1 = []
    db2 = []
    N=len(datos)
    i = 0
    dcosto1 = 0
    while i < N:
        dw1 += [datos[i][0]]
        dw2 += [datos[i][1]]
        db1 += [datos[i][2]]
        db2 += [datos[i][3]]
        dcostox = [(1/n)*sum(dw1),(1/n)*sum(dw2),(1/n)*sum(db1),(1/n)*sum(db2)]
        dcosto1 = np.array([dcostox])
        i = i+1
    return dcosto1


def Grad_C2(Phi, D):
    dcosto2 = np.array([Parcial(Phi, D, 2)])
    return dcosto2


def Gradiente_Cphi(Phi, D):
    gradiente_C1 = Grad_C1(Phi,D)
    gradiente_C2 = Grad_C2(Phi, D)
    DCC = (1/2)*(gradiente_C1+gradiente_C2)
    gradC = [DCC[0,0], DCC[0,1], DCC[0,2],DCC[0,3]]
    return gradC
######################## Funcion gradiente de costos ###########################
# ============================= Tarea 2, Parte 2.b =============================
````

> El conjunto de estas funciones permiten el cálculo de $\sigma(s)''$, $\mathcal{R}(\Phi)$, $C(w_{1}, w_{2}, b_{1}, b_{2})$ y $\nabla C(w_{1}, w_{2}, b_{1}, b_{2})$ respectivamente, además de $\nabla_{(w_{1}, w_{2}, b_{1}, b_{2})}\mathcal{R}^{[orden]}(\Phi)$ para valores de orden 0 (sin derivar) y orden 2 de la realización cuyo resultado fue utilizado para simplificar cálculos.

> ### $3)$ Entrenamiento de $\Phi$ mediante el método del gradiente conjugado.
> Considere los parámetros iniciales de la red: $w_{0}^{1} = 0.5$, $w_{0}^{2} = 1.1$, $b_{0}^{1} = 1.3$, $b_{0}^{2} = 0$. A partir de estos valores iniciales $W_{0} = (w_{0}^{1}, w_{0}^{2}, b_{0}^{1}, b_{0}^{2})$, entrene $\Phi$ con el método de gradiente conjugado:
>
> \begin{equation*} W_{k+1} = W_k - \eta \nabla C(W_k), \quad k = 0, 1, 2, \ldots \end{equation*}

> donde $W_{k} = (w_{k}^{1}, w_{k}^{2}, b_{k}^{1}, b_{k}^{2})^{T}$.
>
> En este caso, considere $\eta = 0.01$ y tres valores distintos de $M$: $M = 100$, $M = 500$, $M = 1000$. Describa sus resultados en detalle, en particular, presente el vector $\Phi_{M} = (w_{M}^{1}, w_{M}^{2}, b_{M}^{1}, b_{M}^{2})$ final.

> Considerando las funciones planteadas para la parte anterior (parte 2.2), se planteó el método del gradiente conjugado para entrenar la red neuronal que se utilizaría. El siguiente código presenta la combinación de estas funciones para la formación del gradiente conjugado respectivo.

````python
# ============================= Tarea 2, Parte 2.b =============================
######################## Funcion gradiente conjugado ###########################
def Gradiente_Conjugado(M, l, Phi0, D, i = 0):
    w1, w2, b1, b2 = Phi0
    for _ in range(M+1):
        grad_w1, grad_w2, grad_b1, grad_b2 = Gradiente_Cphi((w1, w2, b1, b2), D)
        w1 += -l*grad_w1
        w2 += -l*grad_w2
        b1 += -l*grad_b1
        b2 += -l*grad_b2
        #i = i+1
    #print(i)
    return w1, w2, b1, b2
######################## Funcion gradiente conjugado ###########################
# ============================= Tarea 2, Parte 2.b =============================
````

> La función se le asociaría los siguientes comandos y valores constantes con el fin de obtener la realización para distintas cantidades de iteraciones.

````python
# ============================= Tarea 2, Parte 2.b =============================
####################### Ejecucion gradiente conjugado ##########################
nu = 0.01
M1 = 100
M2 = 500
M3 = 1000
W = 0.5, 1.1, 1.3, 0

Caso_M100 = Gradiente_Conjugado(M1, nu, W, xji)
print('Para '+str(M1)+' iteraciones, la red da: ' + str(Caso_M100))
print()
print('De manera matricial, el vector entrenado seria de la forma: \n'+str(np.transpose(np.array([Caso_M100]))))
print()
print()
Caso_M500 = Gradiente_Conjugado(M2, nu, W, xji)
print('Para '+str(M2)+' iteraciones, la red da: ' + str(Caso_M500))
print()
print('De manera matricial, el vector entrenado seria de la forma: \n'+str(np.transpose(np.array([Caso_M500]))))
print()
print()
Caso_M1000 = Gradiente_Conjugado(M3, nu, W, xji)
print('Para '+str(M3)+' iteraciones, la red da: ' + str(Caso_M1000))
print()
print('De manera matricial, el vector entrenado seria de la forma: \n'+str(np.transpose(np.array([Caso_M1000]))))
print()
print()
######################## Funcion gradiente conjugado ###########################
# ============================= Tarea 2, Parte 2.b =============================
````
