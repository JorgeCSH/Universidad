> ## Parte a): "*Modelo Simplificado*"


> Para la parte “$a)$”, se pidió, de manera general, realizar un análisis al sistema de ecuaciones diferenciales tanto como a nivel de solución como gráfico. El sistema estudiado sería el siguiente: \
\
\begin{array}$ \frac{dP}{dt} = -PZ \\ \frac{dZ}{dt} = \gamma PZ \\ \frac{dN}{dt} = (1-\gamma)PZ \end{array}
\
Donde $\gamma$ es un valor constante y P, Z, N, funciones. \
Para el desarrollo también se fueron planteando constantes y arreglos con estas con el fin de optimizar el proceso,  cuadro siguiente es aquel que es utilizado principalmente en esta parte. Finalmente, planteado el enfoque, el desarrollo de parte fue el siguiente: 



````python
# Codigo compartido |parte a|
# Parte a, valores que se utilizan
# Valores Constantes
Gamma = 0.2

# Condiciones iniciales
P0 = 0.8
Z0 = 0.1
N0 = 0.1

# Constantes de estudio
deltat1 = 10
deltat2 = 5
deltat3 = 1
ctes = Gamma
X0 = [P0,Z0,N0]
````

> ### 1. Resolver numéricamente el sistema mediante el método de Euler progresivo para los pasos: $\Delta = 10, \Delta = 5, \Delta = 1$ en el intervalo de tiempo $(t)$ $[0,50]$


>En esta etapa se buscaba resolver el sistema de ecuaciones diferenciales utilizando el método de Euler progresivo, el cual se basa en resolver aproximando una integral en el lado derecho de la EDO a un rectángulo. \
De manera resumida, el método se fundamente en una recurrencia de la forma: \
\
\begin{equation*} y_{n+1} = y_n + hf(x_n,y_n)\end{equation*} \
Siendo que en este caso se trata de un sistema de ecuaciones diferenciales, se recurrirá a aplicar el método “comprimiendo” en una lista los valores a resolver, los cuales serian las funciones planteadas en un inicio. Esto llevaría a la siguiente situación: \
\
$\rightarrow$ $x_n\ $ pasaria a ser $X_n\ $ con $X=[P,N,Z]$ \
$\rightarrow$ $X_{n+1} = X_n + \Delta t_{paso} f( t,X_n)$  
\
Siendo la base del método que se programaría. 
Dado esto, se procedió a programar el método nombrado, el cual resulto en tres cuadros de código principalmente, los cuales serían: \
$I.$  Cuadro en que se define la función. \
$II.$ Cuadro en que se define el método numérico. \
$III.$ Cuadro en que se aplica, soluciona y grafica el método. \
Sin ir más lejos, la resolución quedaría de la siguiente manera:


> $I.\ $Cuadro en que se define la función: \
En este cuadro se define la función del lado derecho. Esta toma tres principales valores: \
\
$\rightarrow$ t: tiempo en que transcurre (no influyente en caso de falta de recurrencia) \
$\rightarrow$ X: Lista que contiene los valores de las funciones $P, N, Z$ \
$\rightarrow$ ctes: Lista que contiene a los valores que actúan como constante, en este caso solo esta el caso $\gamma$ como constante. \
\
En esta función ocurre que se busca un arreglo solución para la derivada de cada ecuación en el sistema, lo que significa que devolvería un valor de la forma “np.array([$\ \cdot\ $])”.\
El funcionamiento de la función se basa en plantear el sistema y las constantes, además de un valor “$X$” inicialmente arbitrario, el cual serviría como forma comprimida de las funciones que serian devueltas como resultado en un arreglo.


````python
# Definir lado derecho de la EDO
def Fderechoeu(t,X,ctes):                                                        # Hablamos de un sistema, por ende hacemos el siguiente orden
    P, N, Z = X                                                                  # Descomprimimos la "constante" que contiene a las funciones a obtener
    Gamma = ctes                                                                 # Descomprimimos la constante que contiene a las demas constantes del sistema (una en este caso)
    derP = -P*Z                                                                  #---|
    derZ = Gamma*P*Z                                                             #   |---->|Derivadas del sistema, especie de   
    derN = (1-Gamma)*P*Z                                                         #---|      denotacion del sistema original
    return np.array([derP, derZ, derN])                                         
````

> $II.\ $Cuadro en que se define el método: \
En este cuadro se define el método utilizado. Para esto se formuló una función que realizase las reiteraciones necesarias “metodoEUR”. La cual recibe los siguientes valores: \
\
$\rightarrow$ F: Función lado derecho \
$\rightarrow$ ti,tf: Tiempo inicial y final en que transcurre el método \
$\rightarrow$ X0: Valor que representa la condición inicial \
$\rightarrow$ deltat: Valor que representa al paso $\Delta t$ utilizado \
$\rightarrow$ ctes: valor de constantes del método \
\
La función tendría el siguiente funcionamiento: Dividir el intervalo principal ($[0, 50]$) en “$N$” subintervalos de largo deltat. Estos subintervalos después se unirían formando un intervalo general “$I$”, para generar un arreglo “$X$” de soluciones y aplicar la iteración, devolviendo así como solución, la tupla “$(I,X)$”.


````python
# Metodo de euler retrogrado
def metodoEUR(F,ti,tf,X0,deltat,ctes):
    N = int((tf-ti)/deltat)                                                      # Dividimos en N sub intervalos de tamaño del peso el intervalo original 
    I = np.linspace(ti,tf,N)                                                     # Los intervalos previos los unimos en un intervalo o conjunto general "I"
    X = np.zeros((N,3))                                                          # Creamos el arreglo solucion
    X[0] = X0                                                                    # Definimos los valores iniciales para iniciar el metodo
    for i in range(N-1):                                                         # -|                         
        X[i+1] = X[i] + deltat*F(I[i],X[i],ctes)                                 # -|---->|Iteracion, se realiza el metodo
    return (I,X)      
````

````python
# Resultados Metodo
I1, X1 = metodoEUR(Fderechoeu,0,50,X0,deltat1,ctes)                              # Resultado Numerico Metodo aplicado
I2, X2 = metodoEUR(Fderechoeu,0,50,X0,deltat2,ctes)                              # Resultado Numerico Metodo aplicado         
I3, X3 = metodoEUR(Fderechoeu,0,50,X0,deltat3,ctes)                              # Resultado Numerico Metodo aplicado


print('Solucion paso = 10:', X1)                                                 # Resultado Numerico (printea, muestra)
print()                                                                          #
print('Solucion paso x = 5', X2)                                                 # Resultado Numerico (printea, muestra)
print()                                                                          #
print('Solucion paso x = 1', X3)                                                 # Resultado Numerico (printea, muestra)
#print(I1, X1)                                                                   # -|
#print()                                                                         #  |
#print(I1, X2)                                                                   #  |--->|Resultado Numerico con intervalo
#print()                                                                         #  |
#print(I3, X3)                                                                   # -|


# Grafico de Resultados                                                          #
plt.figure(figsize=(5,3))                                                        # -|
plt.plot(I1,X1[:,0], label=" P ")                                                #  |
plt.plot(I1,X1[:,1], label=" Z ")                                                #  |
plt.plot(I1,X1[:,2], label=" N ")                                                #  |
plt.xlabel(" Tiempo (t)")                                                        #  |--->|Grafo del caso deltat (paso) = 10  
plt.ylabel(" P, Z, N ")                                                          #  |               
plt.title(" P, Z, N para $\Delta t = 10$ ")                                      #  |               
plt.legend()                                                                     #  |
plt.show()                                                                       # -|                                                                     
print()
plt.figure(figsize=(5,3))                                                        # -|
plt.plot(I2,X2[:,0], label=" P ")                                                #  |
plt.plot(I2,X2[:,1], label=" Z ")                                                #  |
plt.plot(I2,X2[:,2], label=" N ")                                                #  |
plt.xlabel(" Tiempo (t)")                                                        #  |--->|Grafo del caso deltat (paso) = 5 
plt.ylabel(" P, Z, N ")                                                          #  |               
plt.title(" P, Z, N, para $\Delta t = 5$ ")                                      #  |               
plt.legend()                                                                     #  |
plt.show()                                                                       # -|
print()
plt.figure(figsize=(5,3))                                                        # -|
plt.plot(I3,X3[:,0], label=" P ")                                                #  |
plt.plot(I3,X3[:,1], label=" Z ")                                                #  |
plt.plot(I3,X3[:,2], label=" N ")                                                #  |
plt.xlabel(" Tiempo (t)")                                                        #  |--->|Grafo del caso deltat (paso) = 1     
plt.ylabel(" P, Z, N ")                                                          #  |               
plt.title(" P, Z, N, para $\Delta t = 1$ ")                                      #  |               
plt.legend()                                                                     #  |
plt.show()                                                                       # -|
````

````python
# Codigo compartido parte b, VALORES QUE SE UTILIZAN:
# Valores Constantes                
M = 0.1                                                                                                                           
K = 0.2                                                                          
Gamma = 0.3                                                                       

# Condiciones iniciales iniciales
P0 = 0.3                                                                         
Z0 = 0.1                                                                         
N0 = 0.6     

# Funciones del Metodo
Vm = 5
ks = 1
Rm = 2
Lambda = 10 

# Funcion tasa consumo de Nitrogeno por parte del Fitoplancton 
def g(N, Vm, ks):                                                                #
    numerador = N*Vm                                                             # Numerador de funcion informada, cantidad N ponderada por constante
    denominador = ks+N                                                           # Denominador de funcion informada, constante agregada a la cantidad de N
    consumo = numerador/denominador                                              # Se forma fraccion
    return consumo                                                               #

# Funcion tasa de depredacion del Fitoplancton por el Zooplancton 
def h(P, Rm, Lambda):                                                            #
    a = -1*Lambda*P                                                              # Se define el exponente
    f = np.exp(a)                                                                # Se aplica funcion exponencial
    depredacion = Rm*(1-f)                                                       # Se incorpora en formato final
    return depredacion                                                           #

# Constantes de Estudio
deltat1 = 5                                                                      # -|                                           
deltat2 = 1                                                                      #  |--->|Intervalos o pasos a utilizar
deltat3 = 0.1                                                                    # -|
ctes = [M,K,Gamma,Vm,ks,Rm,Lambda]                                               # Lista de constantes a considerar en el metodo 
Y0 = [P0,N0,Z0]                                                                  # Lista que define al valor a aplicar el metodo 
````



> ### 1. Graficar h en funcion de P ($ h(P)$)


> En esta etapa se solcito graficar la funcion $h(P)$ que representaba la tasa de depredacion de fitoplancton con relacion al zooplancton. Para esto se definio inicialmente una funcion encargada de calcular los diferentes valores que tomara esta tasa conforme a la cantidad de fitoplancton que exista, pues es la unica variable que influencia en este proceso (ademas de la constante respectiva). Para lo cual quedaria como:

````python
# Definir la funcion h
def h(PP, Rm, Lambda):                                                           # Función que describe h(P)                                                          
    u = (-1)*Lambda*PP                                                           # Se definen los valores del exponente (u)
    eu = np.exp(u)                                                               # Se plantea la exponencial (exp <=> e)
    h = Rm*(1-eu)                                                                # Se juntan las variables y se retorna la funcion final
    return h
````

>### 2. Aplicar el metodo Runge Kutta 4 para resolver el sistema de EDOs


> Para el siguiente caso, se solicito resolver el sistema de EDOs mediante la utilizacion del metodo numerico Runge-Kutta 4. Este metodo tiene un funcionamiento similar al previamente planteado de Euler progresivo, con la diferencia que se realiza una iteracion con respecto a cuatro ecuaciones (las cuales ocurren al considerar un aumento equivalente a la mitad del anterior en terminos de paso. \
\
Para esta realizacion se sigui el siguiente orden de pasos: \
priemro: formular el lado derecho de la EDO, que se llevo a cabo con la siguiente funcion:

````python
# Lado derecho de la EDO
def FderechoRK(t,Y,ctes): 
    P, N, Z = Y                                                                  # Descompone la constante que contiene a las funciones
    M, K, Gamma, Vm, ks, Rm, Lambda = ctes                                       # Descompone la constante que contene a los valores conocidos (y constantes)
    dP = g(N, Vm, ks)*P-h(P, Rm, Lambda)*Z-M*P                                   # -|
    dZ = Gamma*h(P, Rm, Lambda)*Z-K*Z                                            #  |--->|Sistema de ecuaciones inicial
    dN = -1*g(N, Vm, ks)*P+(1-Gamma)*h(P, Rm, Lambda)*Z+M*P+K*Z                  # -|
    return np.array([dP, dN, dZ])                                                                                        
  
````

> segundo, se genero el metodo en base a una funcion que permitiera iterar estos elementos de la siguiente manera: 

````python
#  Runge-Kutta 4
def metodoRK(F,ti,tf,Y0,deltat,ctes):
    N = int((tf-ti)/deltat)                                                      # Dividimos el intervalo en N sub intervalos de tamaño deltat
    J = np.linspace(ti,tf,N)                                                     # Definimos el nuevo intervalo que contiene estos sub intervalos
    Y = np.zeros((N,3))                                                          # Creamos un arreglo solucion (en este caso matricial)
    Y[0] = Y0                                                                    # Definimos el valor inicial con el fin de iterar
    for i in range(N-1):                                                         # -|
        g1 = F(J[i],Y[i],ctes)                                                   #  |
        g2 = F(J[i]+deltat/2, Y[i]+deltat/2*g1,ctes)                             #  |--->|Iteramos, proceso por el cual se realiza el RK4
        g3 = F(J[i]+deltat/2,Y[i]+deltat/2*g2,ctes)                              #  |
        g4 = F(J[i+1],Y[i]+deltat*g3,ctes)                                       #  |
        Y[i+1] = Y[i] + deltat/6*(g1 + 2*g2 + 2*g3 + g4)                         # -|
    return (J,Y)                                                                 # Tupla final cuyos valore son: J el intervalo e Y funciones comprimidas
````

La segunda, en cambio, es una seccion del codigo encargada de graficar las soluciones solicitadas con el fin de bosquejar el comportamiento de estas variables. \

\
Ahora, tomando los casos en general abarcados en el desarrollo, es importante notar como el tamaño del paso influencia en el comportamiento o suavidad de las curvas. Omitiendo el caso de $\Delta t = 5$ por lo planteado en el desarrollo y considerando los casos $\Delta t = 1, 0.1$, ocurre que mientras mas pequeño sea el paso, la suavidad de las curvas es mayor. La razón debería radicar en el método aplicado, el cual basa su fundamento o desarrollo en un intervalo compuesto por la unión de pequeños subintervalos que correspondan con el tamaño del paso. Esto significaría que para un intervalo constante como el estudiado $[0, 50]$, al disminuir el tamaño del paso, este seria conformado por una mayor cantidad de subintervalos de tamaño menor, lo que significaría que el método numérico utilizado seria aplicado en una mayor cantidad de veces y, por ende, los efectos generados por las tasas de cambio tendrían una menor diferencia o impacto en el grafico final, pues se aplicó el método en mayor medida.  Esto tiene su fundamento matemático el cual se basa en el error de discretización local . Este valor representa al error asociado al método, que dependiendo del valor se estará en condiciones similares a la resolución original. En este caso, y notando que el error nombrado tiene la forma: 
\begin{equation*}
	|d_n| \leq Ch^{p+1}
\end{equation*} \
Considerando que se tratan de las mismas constantes “$C$” y el mismo orden “$p+1$” quedara que dependen del valor del paso ($h$), el cual, disminuirá la diferencia con el error cuando tenga un valor menor, que es lo que ocurre en el caso estudiado, pues para el paso $\Delta t =0.1$ tenemos que tiene un valor menor al paso $\Delta t = 1$, lo que significaría: \
\
\begin{equation*}
	  d_{n0.1}\leq C(0.1)^{p+1} \leq d_{n1}\leq C(1)^{p+1}
\end{equation*}
\
\begin{equation*}
	  \implies d_{n0.1} \leq d_{n1}
\end{equation*}
\
Es decir, el error para el paso $\Delta t = 0.1$ sería menor a aquel en el caso $\Delta t = 1$, lo cual explicaría la suavidad de las curvas.\
\
En el caso del plano de fases, este permite, en base a ordenar como “vectores” o “valores en el plano” a aquellas variables analizadas, permite apreciar el comportamiento de estas entre si durante el proceso.  Es decir, como interaccionan entre si sin necesariamente depender de una variable en común, graficar la interdependencia. Nuevamente utilizaremos el fundamento para el caso $\Delta t = 5$, el cual dado a su comportamiento no aceptable, no sería una buena representación del fenómeno (aunque en este caso, si representa el plano de fases la situación para este caso específico, no el caso general), en cambio, para el caso de la suavidad de las curvas tomando los datos $\Delta t = 1\ y\ 0.1$, ocurrirá que nuevamente representaría a la situación nombrada, pues basta notar que, si se trata de una relación entre las 3 funciones de estudio, entonces al cumplirse la dependencia en los casos anteriores, al interpretarse el caso general debería de mantenerse esta relación, basta incluso notar que el bosquejo del plano de fase es una proyección tridimensional de los casos previamente estudiados. Nuevamente, la suavidad de las curvas seria dependiente del error estudiado anteriormente, pues se trataría de la misma situación y, dado a que en el caso del paso $\Delta t = 0.1$ se encuentra con un menor tamaño, entonces será aquel mas aproximado a la función o grafica original, lo cual es la causa de una curva mas “suave” en este caso.
